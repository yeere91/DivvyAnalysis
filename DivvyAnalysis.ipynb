```python
trips = pd.DataFrame()

for i in [x for x in os.listdir('./data') if 'Trips' in x]:
    ## Figure out quarter by delimiting string
    quarter = i.split("_")[3].split(".")[0]
    df = pd.read_csv('./data/' + i)
    df['quarter'] = quarter
    
    trips = trips.append(df)
```

```python
connection = engine.connect()
from sqlalchemy.sql import text
s = text(
     "SELECT MAX(c)" 
     "FROM"
         "(SELECT COUNT(1) AS c "
          "FROM trips "
          "WHERE to_date(start_time, 'MM/DD/YYYY') < to_date(:d, 'MM/DD/YYYY')"
          "GROUP BY to_date(start_time, 'MM/DD/YYYY HH24:MI:SS')) a"
    )
print connection.execute(s, d='02/07/2017').fetchone()
connection.close()
```

```python
connection = engine.connect()
s = text(
     "SELECT " 
     "COUNT (CASE WHEN tripduration/(60) >= 0 AND tripduration/(60) <= 10 THEN 1 END) AS \"0-10min\","
     "COUNT (CASE WHEN tripduration/(60) > 10 AND tripduration/(60) <= 20 THEN 1 END) AS \"10-20min\","
     "COUNT (CASE WHEN tripduration/(60) > 20 AND tripduration/(60) <= 30 THEN 1 END) AS \"20-30min\","
     "COUNT (CASE WHEN tripduration/(60) > 30 AND tripduration/(60) <= 60 THEN 1 END) AS \"30-60min\","
     "COUNT (CASE WHEN tripduration/(60) > 60 AND tripduration/(60) <= 120 THEN 1 END) AS \"1-2hr\","
     "COUNT (CASE WHEN tripduration/(60) > 120 AND tripduration/(60) <= (60*12) THEN 1 END) AS \"2-12hr\","
     "COUNT (CASE WHEN tripduration/(60) > (60*12) AND tripduration/(60) <= (60*24) THEN 1 END) AS \"12-24hr\" "
     "FROM trips"
    )
results = list(connection.execute(s).fetchone())
connection.close()
```

```python
import plotly.plotly as py
import plotly.graph_objs as go

print results

data = [go.Bar(
            x=['0-10','10-20','20-30','30-60','1-2hr','2-12hr','12-24hr'],
            y=results
    )]

layout = go.Layout(
    yaxis=dict(
        type='log',
        autorange=True
    ))

fig = go.Figure(data=data, layout=layout)
py.iplot(fig)
```
